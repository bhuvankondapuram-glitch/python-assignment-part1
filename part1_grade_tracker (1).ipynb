```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
# Task1
for student in raw_students:
  name = student["name"].strip().title()

  roll = int(student["roll"])

  marks = [int(m) for m in student["marks_str"].split(", ")]

  #Checking if all words in name are alphabetic
  is_valid = all(word.isalpha() for word in name.split())
  validity = "✓ Valid name" if is_valid else "✗ Invalid name"

  #Print profile card
  print("================================")
  print(f"Student : {name} {validity}")
  print(f"Roll No : {roll}")
  print(f"Marks : {marks}")
  print("================================")
```

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
print(f"\nGrade Report for {student_name}")
print("-"* 35)

for i in range(len(subjects)):
  subject = subjects[i]
  mark = marks[i]

  if mark >= 90:
    grade = "A+"
  elif mark >= 80:
    grade = "A"
  elif mark >= 70:
    grade = "B"
  elif mark >= 60:
    grade = "C"
  elif mark < 60:
    grade = "F"

  print(f"{subject:<12} | {mark} | {grade}")

```

```python
total = sum(marks)
average = round(total / len(marks), 2)

highest = max(zip(subjects, marks), key=lambda x: x[1])
lowest = min(zip(subjects, marks), key=lambda x: x[1])

print(f"\nTotal : {total}")
print(f"Average : {average}")
print(f"Highest : {highest[0]} - {highest[1]}")
print(f"Lowest : {lowest[0]} - {lowest[1]}")

```

```python
new_subjects_count = 0

while True:
    subject_input = input("Enter subject name (or 'done' to stop): ")

    if subject_input.lower() == "done":
        break

    marks_input = input(f"Enter marks for {subject_input} (0-100): ")

    # Check 1 — is it even a number?
    if not marks_input.isdigit():
        print("Warning: Please enter a valid number!")
        continue   # skip to next loop iteration

    # Check 2 — is it in range 0-100?
    mark_value = int(marks_input)
    if 0 <= mark_value <= 100:
        subjects.append(subject_input)
        marks.append(mark_value)
        new_subjects_count += 1
        print(f"Added {subject_input} with marks {mark_value}")
    else:
        print("Warning: Marks must be between 0 and 100!")

print(f"\nNew subjects added: {new_subjects_count}")
updated_avg = round(sum(marks) / len(marks), 2)
print(f"Updated average (all subjects): {updated_avg}")
```

```python

class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# Print header
print(f"{'Name':<18}| {'Average':^7} | {'Status'}")
print("-" * 40)

averages = []

for name, marks in class_data:
    avg = round(sum(marks) / len(marks), 2)
    status = "Pass" if avg >= 60 else "Fail"
    averages.append((name, avg, status))
    print(f"{name:<18}| {avg:^7.2f} | {status}")

# Summary
passed    = [s for s in averages if s[2] == "Pass"]
failed    = [s for s in averages if s[2] == "Fail"]
topper    = max(averages, key=lambda x: x[1])
class_avg = round(sum(a[1] for a in averages) / len(averages), 2)

print(f"\nPassed        : {len(passed)} students")
print(f"Failed        : {len(failed)} students")
print(f"Topper        : {topper[0]} with average {topper[1]}")
print(f"Class Average : {class_avg}")
```
